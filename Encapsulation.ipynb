## Enscapsulation

Enscapsilation / Enskapsulasi adalah metode dalam pemograman berorientasi objek yang bertujuan untuk memaketkan atribut - atribut atau method - method dalam class agar bisa menjadi atribut/method yang bersifat private, sehingga dengan metode ini class lain tidak bisa mengakses secara leluasa terhadap atribut/method yang private. Manfaat Enscapsulation :
- Penyembunyian Informasi<br>
  Hal ini mengacu pada pembentukan atribut/method dalam sebuah class menjadi private, sehingga tidak    terlihat di permukaan public, dengan begini jika ada perubhan pada atribut/method yang private tidak akan mengganggu proses dilain class
- Modularitas<br>
  Dengan metode Encapsulaton objek bisa dikelola secara independen, karena objek yang bersifat private ini terpisah dengan bagian interface, jadi bila dilakukan perubahan pada objek yang bersifat private ini tidak akan merusak sistem - sistem yang lain
  
Dalam Python Enscapsulasi bisa dilakukan dengan pembuatan atribut atau method yang bersifat private
##### syntax pembuatan atribut Private
__namaAtribut = IsiVariable
#### syntax pembuatan method Private
__namaMethod()
  

```python
class Pegawai :
    def __init__(self):
        self.__nama=''
        self.__numGaji=0
    def __Gaji(self,bonus):
        print('Gaji Total = ',self.__numGaji + bonus)
    def setNama(self,nama):
        self.__nama=nama
    def setGaji(self,gaji):
        self.__numGaji = gaji
    def getNama(self):
        print(self.__nama)
    def getGaji(self):
        print(self.__numGaji)
    def hitunggaji(self,bonus):
        self.__Gaji(bonus)
        
```

Dalam Pengaksesan Atribut ataupun method private tidak bisa diakses layaknya atribut atau method public<br>
dalam pemanggilan atribut atau method biasa dilakukan dengan<br>
- NamaObjek.Atribut --> untuk pemanggilan atribut
- NamaObjek.Method() --> untuk Pemanggilan Method<br>

yang akan menyebabkan terjadi error sebagai berikut :

<br>
Hal yang harus dilakukan dalam pemangiilan atribut atau method yang bersifat private:<br>
- menggunakan method public perantara untuk mengembalikan nilai atau memuat isi dari atribut atau method yang bersifat private tersebut
- menggunakan syntax khusus dalam pengaksesan Atribut atau Method Private secara langsung:
  NamaObjek._NamaClass.__NamaMethod/NamaAtribut

```python
Aldi = Pegawai()
print('\nMenggunakan Method Perantara')
Aldi.setNama('Reynaldi')
Aldi.setGaji(7000000)
Aldi.getNama()
Aldi.getGaji()
Aldi.hitunggaji(3000000)
print('\nAkses Langsung')
print(Aldi._Pegawai__nama)
print(Aldi._Pegawai__numGaji)
Aldi._Pegawai__Gaji(3000000)
```
